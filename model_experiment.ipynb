```python
corr_matrix = X_train.select_dtypes(include=['number']).corr().abs()
corr_matrix
```

```python
# seminaris codi

X = df.drop(columns=['SalePrice'])
y = df['SalePrice']


# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Find feature pairs with correlation greater than a threshold
threshold = 0.8
high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if corr_matrix.iloc[i, j] > threshold:
            high_corr_pairs.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))

# Display highly correlated pairsF
if high_corr_pairs:
    print("Highly correlated feature pairs:")
    for feat1, feat2, corr in high_corr_pairs:
        print(f"{feat1} and {feat2}: {corr:.4f}")
else:
    print(f"No feature pairs with correlation above {threshold} found.")

# To remove one feature from each highly correlated pair
# (typically the one with lower correlation with target)
features_to_drop = []
for feat1, feat2, _ in high_corr_pairs:
    # Compare correlation with target
    if abs(X[feat1].corr(y)) < abs(X[feat2].corr(y)):
        features_to_drop.append(feat1)
    else:
        features_to_drop.append(feat2)

# Remove duplicates
features_to_drop = list(set(features_to_drop))
print(f"Features to drop due to high correlation: {features_to_drop}")
```

```python
with mlflow.start_run(run_name="one_hot_encoding"):
    mlflow.log_param("original_object_columns", list(remaining_objects))
    mlflow.log_param("drop_first", True)
    
    
    X_train = pd.get_dummies(X_train, columns=remaining_objects, drop_first=True)
    X_test = pd.get_dummies(X_test, columns=remaining_objects, drop_first=True)

    X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

    mlflow.log_metric("final_col_count", X_train.shape[1])
```

```python
import mlflow.sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

selector = SelectKBest(score_func=f_regression, k=50)
X_train_reduced = selector.fit_transform(X_train_scaled, y_train)
X_test_reduced = selector.transform(X_test_scaled)

selected_features = X_train.columns[selector.get_support()]
```

```python
from sklearn.model_selection import KFold, cross_val_score
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

mlflow.end_run()

fill_values = {
    'LotFrontage_median': LotFrontage_median,
    'garage_year_median': garage_year_median,
    'electrical_mode': electrical_mode,
}

def run_experiment(model, name, X_tr, X_te, features_type, selector_obj=None):
    with mlflow.start_run(run_name=f"{name}_{features_type}"):
        
        kf = KFold(n_splits=5, shuffle=True, random_state=42)

        cv_scores = cross_val_score(model, X_tr, y_train, cv=kf, scoring='neg_mean_squared_error')
        cv_rmse_scores = np.sqrt(-cv_scores)
        
        avg_cv_rmse = cv_rmse_scores.mean()
        std_cv_rmse = cv_rmse_scores.std()
        
        model.fit(X_tr, y_train)
        preds = model.predict(X_te)
        test_rmse = np.sqrt(mean_squared_error(y_test, preds))
        
        with open('fill_values.pkl', 'wb') as f:
            pickle.dump(fill_values, f)
        with open('scaler.pkl', 'wb') as f:
            pickle.dump(scaler, f)
        
        mlflow.log_artifact('fill_values.pkl')
        mlflow.log_artifact('scaler.pkl')
        
        if selector_obj is not None:
            with open('selector.pkl', 'wb') as f:
                pickle.dump(selector_obj, f)
            mlflow.log_artifact('selector.pkl')

        mlflow.log_param("model_name", name)
        mlflow.log_param("features_count", X_tr.shape[1])
        
        if hasattr(model, 'max_depth'):
            mlflow.log_param("max_depth", model.max_depth)
            
        mlflow.log_metric("cv_rmse_mean", avg_cv_rmse)
        mlflow.log_metric("cv_rmse_std", std_cv_rmse)
        mlflow.log_metric("test_rmse", test_rmse)
        
        mlflow.sklearn.log_model(model, "model", registered_model_name=name)

run_experiment(LinearRegression(), "Linear_Reg_KFold", X_train_scaled, X_test_scaled, "all_features")
run_experiment(DecisionTreeRegressor(max_depth=30, random_state=42), "DT_Overfit_KFold", X_train_scaled, X_test_scaled, "all_features")
run_experiment(DecisionTreeRegressor(max_depth=10, random_state=42), "DT_Tuned_KFold", X_train_reduced, X_test_reduced, "reduced_features", selector_obj=selector)
```
